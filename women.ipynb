```python
df['Rating'].value_counts()
```

```python
def a (rating):
    if rating >=4:
        return 'positive'
    elif rating ==3:
        return 'neutral'
    else:
        return 'negative'
    
```

```python
df['Sentiment']=df['Rating'].apply(a)
```

```python
df.head(5)
```

```python
correlation_matrix = df[['Rating', 'Recommended IND', 'Positive Feedback Count']].corr()

print(correlation_matrix)
```

```python
# Cross-tabulation of Rating and Recommended IND
rating_recommendation_crosstab = pd.crosstab(df['Rating'], df['Recommended IND'])

print(rating_recommendation_crosstab)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=20)
plt.title('Age Distribution of Reviewers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sentiment_count = df['Sentiment'].value_counts()
```

```python
plt.figure(figsize=(7, 10))
plt.pie(sentiment_count, labels=sentiment_count.index, autopct='%1.1f%%', startangle=140)
plt.title('Sentiment Distribution of Reviewers')
plt.legend(title="Sentiment :")
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.scatterplot(x='Rating', y='Positive Feedback Count', data=df, alpha=0.6)
plt.title('Rating vs. Positive Feedback Count')
plt.xlabel('Rating')
plt.ylabel('Positive Feedback Count')
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.barplot(x='Rating', y='Class Name', data=df)
plt.title('Rating for Cloth')
plt.xlabel('Rating')
plt.ylabel('cloth name')
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.boxplot(x='Age', y='Class Name', data=df)
plt.xlabel('Rating')
plt.ylabel('cloth name')
plt.show()
```

```python
avg_rating_by_class = df.groupby('Class Name')['Rating'].mean().reset_index()
```

```python
plt.figure(figsize=(8, 8))
sns.barplot(x='Class Name', y='Rating', data=avg_rating_by_class, palette='viridis')
plt.title('Average Rating by Class Name')
plt.xlabel('Class Name')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)
plt.show()
```

```python
labels = avg_rating_by_class['Class Name']
sizes = avg_rating_by_class['Rating']
```

```python
plt.figure(figsize=(10, 10))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("viridis", len(labels)))
plt.title('Average Rating by Cloth Name')
plt.axis('equal')  
plt.show()
```
